```python
# Topic 9: Working with JSON files

import json

student1 = {
    "id" : 1,
    "name" : "Ravi",
    "age" : 20
}

with open("student1.json", "w") as file:
    json.dump(student1,file,indent=4)
    
```

```python
import json

with open("student1", "r") as file:
    data = file.read()
print(data)
print(data["name"])
```

```python
import json

with open("student1.json", "r") as file:
    data = json.load(file)
print(data)
print(data["name"])
```
