```python
# file handling basics

with open("notes.txt", "r") as file:
    content = file.read()
print(content)
```

```python
# •	Create a text file with two lines and read it using Python.

with open("sample.txt", "r") as file:
    data = file.read()
print(data)
```

```python
# •	Change the mode from r to w in a separate example and explain what w means.

with open("sample.txt", "w") as file:
    data = file.read()
print(data)
```
