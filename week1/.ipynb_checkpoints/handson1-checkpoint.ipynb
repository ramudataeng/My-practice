```python
# •	Read the CSV with DictReader.

import csv

with open("handson1.csv", "r") as file:
    data = csv.DictReader(file)
    for row in data:
        print(row)


```

```python
# •	Look at the header names and convert them into a cleaner style such as lowercase with underscores.

import csv

with open("handson1.csv", "r") as file:
    data = csv.DictReader(file)
    original_headers = data.fieldnames
    print(original_headers)
    cleaned_headers = [h.strip().lower().replace(" ","_") for h in original_headers]
    print(cleaned_headers)
```

```python
# remove nulls and empty spaces

import csv

with open("handson1.csv" ,"r") as file:
    data = csv.DictReader(file)
    
    for row in data:
        for key in row:
            row[key] = row[key] or 0
        print(row)
```

```python
# how to convert to json file after performing all at a time


import csv
import json

input_file = "handson1.csv"
output_file = "cleaned_data.json"

cleaned_data = []

with open(input_file, "r") as file:
    data = csv.DictReader(file)
    original_headers = data.fieldnames
    cleaned_headers = [h.strip().lower().replace(" ","_") for h in original_headers]
    print(original_headers)
    print(cleaned_headers)
   
    for row in data:
        new_row = {}
        for old_key, new_key in zip(original_headers,cleaned_headers):
            new_row[new_key] = row[old_key] or 0
            
        cleaned_data.append(new_row)

    with open(output_file, "w") as file:
        json.dump(cleaned_data, file, indent =4)
        
        
    
   
```

```python
# read the json file

import json

with open("cleaned_data.json", "r") as file:
    data = json.load(file)
    print(data)
```
