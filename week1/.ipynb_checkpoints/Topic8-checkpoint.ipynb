```python
# Topic 8: Working with CSV files

import csv

with open("students.csv", "r") as file:
    data = file.read()
print(data)
```

```python
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
        
```

```python
import csv
with open("students.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)
        
```

```python
# •	Ask the student to print only row["name"] inside the loop.

import csv
with open("students.csv", "r") as file:
    data = csv.DictReader(file)
    for row in data:
        print(row["name"])
        
```

```python
# using csv.reader
import csv
with open("students.csv", "r") as file:
    data = csv.reader(file)
    for row in data:
        print(row[1:3])
```
