```python
# Dictionary

student = {
    "name": "Asha",
    "Age": 14,
    "Grade": "10th"
}
print(student["Grade"])
```

```python
#lists

numbers = [1,2,3,4,5]
Largest = max(numbers)
print(Largest)
```

```python
num = [1,2,3,4,5]
num.reverse()
print(num)
```

```python
# tuple
num = (1,2,3,4,5)
print(num[3])
print(num[-1])
```

```python
# sets
s = {1,2,3}
s.add(4)
print(s)
s.remove(3)
print(s)
```
