```python
# Functions and reusable code

def clean_age(age):
    if age == "":
        return 0
    return int(age)
value = clean_age("35")
print(value)
```

```python
# •	Write a function that receives a product name and returns it in uppercase.
def product_name(name):
    return name.upper()
value = product_name("soap")
print(value)
```

```python
def product_name(name):
    return name.upper()
name = input()
value = product_name(name)
print(value)
```
